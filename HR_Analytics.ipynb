```python
import pandas as pd
df = pd.read_csv('HR_Analytics.csv')
df.head()
```

```python
import matplotlib.pyplot as plt
df.groupby('Department')['EducationField'].count().plot(kind='pie')
plt.title('HR Analytics')
plt.xlabel('Department')
plt.ylabel('Count of EducationField')
plt.show()
```

```python
import matplotlib.pyplot as plt
df.groupby('Department')['DailyRate'].count().plot(kind='bar')
plt.title('HR Analytics')
plt.xlabel('Department')
plt.ylabel('DailyRate')
plt.show()
```
